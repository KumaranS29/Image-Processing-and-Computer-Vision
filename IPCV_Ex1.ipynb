```python
#Splitting and merging color channels
import cv2
import matplotlib.pyplot as plt
flower=cv2.imread(r"C:\Users\lenovo\Downloads\flower.jpg",cv2.IMREAD_COLOR)
b,g,r=cv2.split(flower)
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(r,cmap="gray")
plt.title("Red Channel")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(g,cmap="gray")
plt.title("Green Channel")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(b,cmap="gray")
plt.title("Blue Channel")
plt.axis("off")
plt.show()
```

```python
#Changing BGR to HSV color space
```

```python
import cv2
import matplotlib.pyplot as plt
flower=cv2.imread(r"C:\Users\lenovo\Downloads\flower.jpg",cv2.IMREAD_COLOR)
hsv = cv2.cvtColor(flower, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)

plt.figure(figsize=[20,5])

plt.subplot(1,4,1)
plt.imshow(h, cmap="gray")
plt.title("H Channel")
plt.axis("off")

plt.subplot(1,4,2)
plt.imshow(s, cmap="gray")
plt.title("S Channel")
plt.axis("off")

plt.subplot(1,4,3)
plt.imshow(v, cmap="gray")
plt.title("V Channel")
plt.axis("off")

plt.subplot(1,4,4)
plt.imshow(cv2.cvtColor(flower, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis("off")

plt.show()
```

```python
#Display Hue , Saturation and Value (HSV) Channels  
```

```python
import cv2
import matplotlib.pyplot as plt
flower=cv2.imread(r"C:\Users\lenovo\Downloads\flower.jpg",cv2.IMREAD_COLOR)
hsv=cv2.cvtColor(flower,cv2.COLOR_BGR2HSV)
h,s,v=cv2.split(hsv)
h_new = h+30
flower_merged=cv2.merge((h_new,s,v))
flower_rgb=cv2.cvtColor(flower_merged,cv2.COLOR_HSV2RGB)
h,s,v=cv2.split(flower_merged)
plt.figure(figsize=[20,5])

plt.subplot(1,4,1)
plt.imshow(h,cmap="grey")
plt.title("H+30 Channel")
plt.axis("off")

plt.subplot(1,4,2)
plt.imshow(s,cmap="grey")
plt.title("S Channel")
plt.axis("off")

plt.subplot(1,4,3)
plt.imshow(v,cmap="grey")
plt.title("V Channel")
plt.axis("off")

plt.subplot(1,4,4)
plt.imshow(flower_rgb)
plt.title("Merged Image")
plt.axis("off")
plt.show()
```
