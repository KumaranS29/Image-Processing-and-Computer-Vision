```python
#Read and Display Image
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
flower = cv2.imread(r"C:\Users\lenovo\Downloads\rose.webp")

# Convert BGR (OpenCV default) to RGB (Matplotlib default)
flower1 = cv2.cvtColor(flower, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(flower1)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
#Cropping
```

```python
#Resizing
```

```python
import cv2
flower = cv2.imread(r"C:\Users\lenovo\Downloads\rose.webp")
resized = cv2.resize(flower, (300, 300))
flower1 = cv2.cvtColor(resized, cv2.COLOR_BGR2RGB)
plt.imshow(flower1)
plt.title("Resized Image")
plt.axis("off")
plt.show()
```

```python
#Translation
```

```python
import cv2
flower = cv2.imread(r"C:\Users\lenovo\Downloads\rose.webp")
rows, cols = flower.shape[:2]
M = np.float32([[1, 0, 50],
[0, 1, 30]])
translated = cv2.warpAffine(flower, M, (cols, rows))
flower1 = cv2.cvtColor(translated, cv2.COLOR_BGR2RGB)
plt.imshow(flower1)
plt.title("Translated Image")
plt.axis("off")
plt.show()
```

```python
#Rotation
```

```python
import cv2
flower = cv2.imread(r"C:\Users\lenovo\Downloads\rose.webp")
rows, cols = flower.shape[:2]
center = (cols // 2, rows // 2)
M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(flower, M, (cols, rows))
flower1 = cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)
plt.imshow(flower1)
plt.title("Rotated Image")
plt.axis("off")
plt.show()
```

```python
#Scaling
```

```python
import cv2
flower = cv2.imread(r"C:\Users\lenovo\Downloads\rose.webp")
scaled = cv2.resize(flower, None, fx=1.5, fy=1.5)
flower1 = cv2.cvtColor(scaled, cv2.COLOR_BGR2RGB)
plt.imshow(flower1)
plt.title("Scaled Image")
plt.axis("off")
plt.show()
```

```python
#Shearing
```

```python
import cv2
flower = cv2.imread(r"C:\Users\lenovo\Downloads\rose.webp")
rows, cols = flower.shape[:2]
M = np.float32([[1, 0.5, 0],
               [0, 1, 0]])
sheared = cv2.warpAffine(flower, M, (int(cols * 1.5), rows))
flower1 = cv2.cvtColor(sheared, cv2.COLOR_BGR2RGB)
plt.imshow(flower1)
plt.title("Sheared Image")
plt.axis("off")
plt.show()
```

```python
#Reflection (flip)
```

```python
import cv2
flower = cv2.imread(r"C:\Users\lenovo\Downloads\rose.webp")
flipped = cv2.flip(flower, 1)
flower1 = cv2.cvtColor(flipped, cv2.COLOR_BGR2RGB)
plt.imshow(flower1)
plt.title("Reflected Image")
plt.axis("off")
plt.show()
```

```python
#Perspective Transformation
```

```python
import cv2
flower = cv2.imread(r"C:\Users\lenovo\Downloads\rose.webp")
rows, cols = flower.shape[:2]

pts1 = np.float32([[50, 50],
[cols-50, 50],
[50, rows-50],
[cols-50, rows-50]])

pts2 = np.float32([[10, 100],
[cols-100, 50],
[100, rows-10],
[cols-50, rows-50]])

M = cv2.getPerspectiveTransform(pts1, pts2)
perspective = cv2.warpPerspective(flower, M, (cols, rows))

flowerl = cv2.cvtColor(perspective, cv2.COLOR_BGR2RGB)
plt.imshow(flowerl)
plt.title("Perspective Transformation")
plt.axis("off")
plt.show()
```
