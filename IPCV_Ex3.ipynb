```python
#Sobel Edge Detection
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread(r"C:\Users\lenovo\Downloads\rose.webp")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
Ix = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
Iy = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
edges = cv2.magnitude(Ix, Iy)
edges = np.uint8(255 * edges / np.max(edges))
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Original Image")
plt.subplot(122)
plt.imshow(edges, cmap='gray')
plt.axis("off")
plt.title("Sobel Edges")
plt.show()
```

```python
#Prewitt Edge Detection
```

```python
import cv2
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt
img = cv2.imread(r"C:\Users\lenovo\Downloads\rose.webp")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
px = np.array([[-1,0,1],[-1,0,1],[-1,0,1]], dtype=float)
py = np.array([[-1,-1,-1],[0,0,0],[1,1,1]], dtype=float)
Ix = ndimage.convolve(gray.astype(float), px)
Iy = ndimage.convolve(gray.astype(float), py)
edges = np.hypot(Ix, Iy)
edges = np.uint8(255 * edges/np.max(edges))
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Original Image")
plt.subplot(122)
plt.imshow(edges, cmap='gray')
plt.axis("off")
plt.title("Prewitt Edges")
plt.show()
```

```python
#Roberts Cross Edge Detection
```

```python
import cv2
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt
img = cv2.imread(r"C:\Users\lenovo\Downloads\rose.webp")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
rx = np.array([[1,0], [0,-1]], dtype=float)
ry= np.array([[0,1],[-1,0]], dtype=float)
Ix = ndimage.convolve(gray.astype(float), rx)
Iy = ndimage.convolve(gray.astype(float), ry)
edges = np.hypot(Ix, Iy)
edges = np.uint8(255 * edges/np.max(edges))
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Original Image")
plt.subplot(122)
plt.imshow(edges, cmap='gray')
plt.axis("off")
plt.title("Roberts Edges")
plt.show()
```

```python
#Line Detection Using Hough Transform
```

```python
import cv2
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt
img = cv2.imread(r"C:\Users\lenovo\Downloads\house.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.GaussianBlur(gray, (5,5), 0)
edges = cv2.Canny(gray, 80, 150)
lines = cv2.HoughLinesP(edges, 1, np.pi/180, threshold=50, minLineLength=10,maxLineGap=10)
line_img = img.copy()
if lines is not None:
    for x1, y1, x2, y2 in lines[:,0]:
        cv2.line(line_img, (x1, y1), (x2, y2), (0,255,0), 2)

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Original Image")
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(line_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Hough Transform")
plt.show()
```

```python
#Harris Corner Detection
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread(r"C:\Users\lenovo\Downloads\Chess_Board.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY).astype(np.float32) / 255.0

Ix = ndimage.sobel(gray, axis=1)
Iy = ndimage.sobel(gray, axis=0)

sigma = 1.5
Ix2 = ndimage.gaussian_filter(Ix ** 2, sigma)
Iy2 = ndimage.gaussian_filter(Iy ** 2, sigma)

Ixy = ndimage.gaussian_filter(Ix * Iy, sigma)

k = 0.05
det = Ix2 * Iy2 - Ixy ** 2
trace = Ix2 + Iy2
R = det - k * trace ** 2

R_thresh = np.where(R > 0.15 * R.max(), R, 0)
R_max = ndimage.maximum_filter(R_thresh, size=10)
local_max = (R_thresh == R_max) & (R_thresh > 0)
corners = np.argwhere(local_max)

plt.figure(figsize=(8, 8))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.scatter(corners[:, 1], corners[:, 0], c='red', s=60, marker='x', linewidths=1.5)
plt.title(f"Harris Corner ({len(corners)} detected)")
plt.axis('off')
plt.tight_layout()
plt.show()
```
